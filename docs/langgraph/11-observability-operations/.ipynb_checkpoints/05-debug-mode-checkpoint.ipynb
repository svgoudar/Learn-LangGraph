## Debug Mode

**Debug Mode** in LangGraph is a **developer observability and diagnostics framework** that exposes internal execution behavior of a graph — including **state transitions, node execution order, decisions, failures, retries, and performance characteristics** — to help engineers **build, validate, and harden** LLM workflows.

It transforms a LangGraph application from a black box into an **inspectable, deterministic system**.

---

### **1. Why Debug Mode Exists**

LLM workflows involve:

* Non-deterministic model behavior
* Complex control flow (loops, branches, parallelism)
* External tools & APIs
* Persistent state and memory

Without debugging, failures are **opaque** and **unreproducible**.

**Debug Mode provides:**

| Capability            | Purpose                      |
| --------------------- | ---------------------------- |
| Execution Trace       | Understand *what happened*   |
| State Inspection      | Understand *why it happened* |
| Determinism Replay    | Reproduce issues             |
| Performance Profiling | Optimize cost & latency      |
| Failure Localization  | Identify faulty nodes        |

---

### **2. What Debug Mode Exposes**

When debug is enabled, LangGraph records:

| Layer           | Information                         |
| --------------- | ----------------------------------- |
| Node Execution  | order, duration, inputs, outputs    |
| State Evolution | full state before & after each node |
| Edge Decisions  | why a branch was chosen             |
| LLM Calls       | prompts, responses, tokens          |
| Tool Calls      | inputs, outputs, errors             |
| Retries         | attempts, backoff, failures         |
| Timing          | latency per node                    |
| Errors          | exception stacktraces               |
| Interrupts      | human intervention events           |

---

### **3. Enabling Debug Mode**

```python
result = graph.invoke(
    input,
    config={"debug": True}
)
```

or globally:

```python
import os
os.environ["LANGGRAPH_DEBUG"] = "true"
```

---

### **4. Execution Trace Example**

```text
[DEBUG] Enter Node: planner
State Before: {goal: "...", step: 0}

[DEBUG] Exit Node: planner
State After: {plan: "...", step: 1}

[DEBUG] Routing Decision: executor

[DEBUG] Enter Node: executor
LLM Prompt: ...
LLM Response: ...
Latency: 1.34s
Tokens: 623
```

This allows **line-by-line reconstruction** of the full workflow.

---

### **5. Debugging Cyclic Graphs**

In cyclic systems, Debug Mode is essential for detecting:

* Infinite loops
* Incorrect termination conditions
* State corruption
* Unbounded memory growth

```python
graph.invoke(
    input,
    config={"debug": True, "recursion_limit": 20}
)
```

---

### **6. Debug Mode + Visualization**

Debug traces align directly with visual graphs:

```python
display(graph.get_graph().draw_mermaid())
```

You can correlate:

> **visual edge → execution trace → state change**

---

### **7. Failure Diagnosis Workflow**

1. Enable Debug Mode
2. Run failing input
3. Inspect:

   * Node where failure occurred
   * State before failure
   * Last successful transition
4. Replay execution using checkpoints
5. Apply fix and re-run

---

### **8. Debug Mode vs Production Mode**

| Feature       | Debug    | Production |
| ------------- | -------- | ---------- |
| State logging | Full     | Minimal    |
| Tracing       | Enabled  | Selective  |
| Performance   | Slower   | Optimized  |
| Observability | Maximum  | Controlled |
| Cost tracking | Detailed | Aggregated |

**Production best practice:**
Use Debug Mode in **staging**, not in **high-throughput production**.

---

### **9. Debug Mode as a Safety System**

Debug Mode provides:

* Auditability
* Compliance evidence
* Reproducibility
* Root-cause analysis
* Safe iteration of autonomous systems

Without it, complex LangGraph systems are **unsafe to operate at scale**.

---

### **10. Mental Model**

Think of Debug Mode as:

> **A flight recorder for LLM workflows**

It records **every decision, state change, and event**, enabling precise diagnosis and continuous improvement.



### Demonstration

````python
# Single-cell LangGraph Debug Mode Demo

from typing import TypedDict
from langgraph.graph import StateGraph, END

# ---------------------------
# 1. Define State
# ---------------------------
class State(TypedDict):
    counter: int

# ---------------------------
# 2. Define Nodes
# ---------------------------
def increment(state: State) -> State:
    return {"counter": state["counter"] + 1}

def check(state: State):
    if state["counter"] >= 3:
        return END
    return "increment"

# ---------------------------
# 3. Build Cyclic Graph
# ---------------------------
builder = StateGraph(State)

builder.add_node("increment", increment)
builder.set_entry_point("increment")

builder.add_conditional_edges(
    "increment",
    check,
    {"increment": "increment", END: END}
)

graph = builder.compile()

# ---------------------------
# 4. Visualize Graph
# ---------------------------
from IPython.display import display
display(graph.get_graph().draw_mermaid())

# ---------------------------
# 5. Run with Debug Mode
# ---------------------------
result = graph.invoke(
    {"counter": 0},
    config={"debug": True, "recursion_limit": 10}
)

print("\nFinal Result:", result)
````

## **Agent Protocol in LangGraph**

An **Agent Protocol** in LangGraph is the **formal communication and coordination framework** that governs how multiple autonomous agents exchange information, make decisions, delegate tasks, and converge toward a shared objective using the graph’s state and control-flow mechanisms.

In essence:

> **Agent Protocol = Rules + State + Message Flow + Control Logic**

---

### **1. Why Agent Protocols Are Necessary**

Single agents cannot robustly handle complex, uncertain tasks.
Multi-agent systems require:

| Requirement     | Purpose                      |
| --------------- | ---------------------------- |
| Role separation | Specialization of reasoning  |
| Communication   | Sharing intermediate results |
| Coordination    | Prevent conflicts            |
| Delegation      | Task distribution            |
| Verification    | Error reduction              |
| Convergence     | Reach final answer           |

LangGraph implements these guarantees **natively through its state machine design**.

---

### **2. Core Components of an Agent Protocol**

| Component             | Role                       |
| --------------------- | -------------------------- |
| **Agents**            | Autonomous reasoning units |
| **Shared State**      | Communication medium       |
| **Message Objects**   | Structured data exchanged  |
| **Control Rules**     | Who speaks next            |
| **Routing Logic**     | Which agent executes       |
| **Termination Rules** | When protocol stops        |

---

### **3. Message-Based Communication Model**

LangGraph agents **do not call each other directly**.
They communicate via **shared state transitions**.

```python
class State(TypedDict):
    messages: list[dict]
    task_status: str
```

Each agent **reads state → writes new message → updates status**.

Example message:

```python
{"sender": "ResearchAgent", "content": "Collected 5 papers on topic X"}
```

---

### **4. Basic Agent Protocol Workflow**

```
Supervisor → ResearchAgent → ComputeAgent → ReviewerAgent → Supervisor → END
```

Each transition is encoded as graph edges.

```python
builder.add_edge("supervisor", "research")
builder.add_edge("research", "compute")
builder.add_edge("compute", "review")
builder.add_edge("review", "supervisor")
```

This creates a **closed-loop coordination protocol**.

---

### **5. Role-Based Protocol Design**

| Agent         | Responsibility          |
| ------------- | ----------------------- |
| Supervisor    | Task planning & routing |
| ResearchAgent | Information gathering   |
| ComputeAgent  | Calculation & synthesis |
| ReviewerAgent | Verification & critique |

Agents follow **role-specific policies**.

```python
def research_agent(state):
    state["messages"].append({"sender": "research", "content": "Found data"})
    return state
```

---

### **6. Conditional Routing (Decision Authority)**

Supervisor chooses next agent dynamically.

```python
def route(state):
    if state["task_status"] == "need_review":
        return "review"
    if state["task_status"] == "complete":
        return END
    return "research"
```

```python
builder.add_conditional_edges("supervisor", route, {
    "research": "research",
    "review": "review",
    END: END
})
```

This makes the protocol **adaptive**.

---

### **7. Protocol Variants**

| Protocol Pattern  | Purpose                        |
| ----------------- | ------------------------------ |

### **9. Minimal Multi-Agent Protocol Example**

```python
from langgraph.graph import StateGraph, END
from typing import TypedDict, List

class State(TypedDict):
    messages: List[str]
    done: bool

def supervisor(state):
    return state

def worker(state):
    state["messages"].append("work done")
    state["done"] = True
    return state

builder = StateGraph(State)
builder.add_node("supervisor", supervisor)
builder.add_node("worker", worker)

builder.set_entry_point("supervisor")
builder.add_edge("supervisor", "worker")
builder.add_edge("worker", "supervisor")

builder.add_conditional_edges("supervisor",
    lambda s: END if s["done"] else "worker",
    {"worker": "worker", END: END}
)

graph = builder.compile()
print(graph.invoke({"messages": [], "done": False}))
```

### **10. Mental Model**

> **LangGraph Agent Protocol = Distributed reasoning system encoded as a deterministic state machine**

It combines:

* Agent autonomy
* Formal coordination
* State-based communication
* Production safety guarantees


### Demonstration

````python
from typing import TypedDict, List
from langgraph.graph import StateGraph, END

# -------------------------
# 1. Shared State Schema
# -------------------------

class State(TypedDict):
    messages: List[str]
    done: bool

# -------------------------
# 2. Agents
# -------------------------

def supervisor(state: State):
    state["messages"].append("Supervisor: checking")
    return state

def worker(state: State):
    state["messages"].append("Worker: finished")
    state["done"] = True
    return state

def reviewer(state: State):
    state["messages"].append("Reviewer: approved")
    return state

# -------------------------
# 3. Router
# -------------------------

def route(state: State):
    if state["done"]:
        return END
    return "worker"

# -------------------------
# 4. Build Graph
# -------------------------

builder = StateGraph(State)

builder.add_node("supervisor", supervisor)
builder.add_node("worker", worker)
builder.add_node("reviewer", reviewer)

builder.set_entry_point("supervisor")

builder.add_edge("worker", "reviewer")
builder.add_edge("reviewer", "supervisor")

builder.add_conditional_edges(
    "supervisor",
    route,
    {"worker": "worker", END: END}
)

graph = builder.compile()

# -------------------------
# 5. Run
# -------------------------

result = graph.invoke(
    {"messages": [], "done": False},
    config={"recursion_limit": 10}
)

print(result)
````

## Compensating Action

A **compensating action** in LangGraph is a **corrective operation** that reverses or mitigates the effects of a previously executed step when the workflow fails, is interrupted, or must be rolled back.
It enables **transaction-like guarantees** in distributed, long-running LLM workflows.

---

### **1. Motivation: Why Compensating Actions Are Needed**

LLM systems often execute **non-atomic operations**:

* API calls
* Database writes
* File uploads
* External service actions
* Payments / reservations / bookings

Once executed, these actions **cannot simply be ‚Äúundone‚Äù**.
Compensating actions provide **logical rollback**.

| Traditional DB     | LangGraph               |
| ------------------ | ----------------------- |
| ACID rollback      | Saga-style compensation |
| Short transactions | Long-running workflows  |
| Centralized        | Distributed & async     |

LangGraph implements this using **explicit compensation logic encoded into the graph**.

---

### **2. Core Conceptual Model**

```
Step A  ‚Üí  Step B  ‚Üí  Step C
   ‚Üì        ‚Üì        ‚Üì
Undo A   Undo B   Undo C
```

If **Step C fails**, the system executes:

```
Undo B ‚Üí Undo A
```

This is the **Saga Pattern** implemented as a **graph**.

---

### **3. How LangGraph Implements Compensation**

Compensation is modeled using:

* **Error nodes**
* **Fallback edges**
* **State-based routing**
* **Checkpoint replay**
* **Explicit rollback nodes**

Each critical step has an associated **compensation node**.

---

### **4. Minimal Working Example**

### **State**

```python
class State(TypedDict):
    order_created: bool
    payment_charged: bool
    shipping_scheduled: bool
    error: str | None
```

### **Forward Actions**

```python
def create_order(state):
    return {"order_created": True}

def charge_payment(state):
    return {"payment_charged": True}

def schedule_shipping(state):
    raise Exception("Shipping API failed")
```

### **Compensating Actions**

```python
def refund_payment(state):
    print("Refunding payment")
    return {"payment_charged": False}

def cancel_order(state):
    print("Cancelling order")
    return {"order_created": False}
```

---

### **Graph Construction**

```python
from langgraph.graph import StateGraph, END

builder = StateGraph(State)

builder.add_node("create_order", create_order)
builder.add_node("charge_payment", charge_payment)
builder.add_node("schedule_shipping", schedule_shipping)

builder.add_node("refund", refund_payment)
builder.add_node("cancel", cancel_order)

builder.set_entry_point("create_order")
builder.add_edge("create_order", "charge_payment")
builder.add_edge("charge_payment", "schedule_shipping")

# Error handling & compensation
builder.add_edge("schedule_shipping", "refund")
builder.add_edge("refund", "cancel")
builder.add_edge("cancel", END)

graph = builder.compile()
```

---

### **5. Execution Semantics**

When `schedule_shipping` fails:

1. The graph routes to **refund**
2. Refund undoes payment
3. Cancel undoes order creation
4. Workflow terminates safely

This preserves **system consistency**.

---

### **6. Production Design Pattern**

| Stage    | Action         | Compensation    |
| -------- | -------------- | --------------- |
| Order    | Create order   | Cancel order    |
| Payment  | Charge payment | Refund payment  |
| Shipping | Book shipment  | Cancel shipment |

This is called the **Saga Workflow Pattern**.

---

### **7. When to Use Compensating Actions**

Use when your workflow:

* Touches external systems
* Performs irreversible operations
* Runs longer than seconds
* Must guarantee consistency
* Must tolerate partial failure

---

### **8. Safety & Reliability Controls**

| Control        | Purpose                      |
| -------------- | ---------------------------- |
| Checkpointing  | Resume or rollback           |
| Idempotency    | Avoid duplicate effects      |
| Retry policy   | Recover transient errors     |
| Audit log      | Full traceability            |
| Human approval | For destructive compensation |

---

### **9. Conceptual Summary**

> **Compensating actions transform LangGraph from a workflow engine into a transaction coordinator for AI systems.**

They allow you to build **fault-tolerant, distributed, autonomous AI pipelines** that remain consistent even under failure.


### Demonstration

````python
from typing import TypedDict
from langgraph.graph import StateGraph, END

# -------------------------------
# 1. Shared State
# -------------------------------

class State(TypedDict):
    order_created: bool
    payment_charged: bool
    shipping_scheduled: bool
    failed: bool

# -------------------------------
# 2. Forward Actions
# -------------------------------

def create_order(state: State):
    print("Order created")
    return {"order_created": True}

def charge_payment(state: State):
    print("Payment charged")
    return {"payment_charged": True}

def schedule_shipping(state: State):
    print("Shipping failed ‚ùå")
    return {"failed": True}

# -------------------------------
# 3. Compensating Actions
# -------------------------------

def refund_payment(state: State):
    print("Refunding payment üîÑ")
    return {"payment_charged": False}

def cancel_order(state: State):
    print("Cancelling order üîÑ")
    return {"order_created": False}

# -------------------------------
# 4. Router Logic
# -------------------------------

def route_after_shipping(state: State):
    if state["failed"]:
        return "refund"
    return END

# -------------------------------
# 5. Build Graph
# -------------------------------

builder = StateGraph(State)

builder.add_node("create_order", create_order)
builder.add_node("charge_payment", charge_payment)
builder.add_node("schedule_shipping", schedule_shipping)
builder.add_node("refund", refund_payment)
builder.add_node("cancel", cancel_order)

builder.set_entry_point("create_order")

builder.add_edge("create_order", "charge_payment")
builder.add_edge("charge_payment", "schedule_shipping")

builder.add_conditional_edges("schedule_shipping", route_after_shipping, {
    "refund": "refund",
    END: END
})

builder.add_edge("refund", "cancel")
builder.add_edge("cancel", END)

graph = builder.compile()

# -------------------------------
# 6. Execute
# -------------------------------

initial_state = {
    "order_created": False,
    "payment_charged": False,
    "shipping_scheduled": False,
    "failed": False
}

result = graph.invoke(initial_state)
print("\nFinal State:", result)
````
